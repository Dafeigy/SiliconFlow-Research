### 定义余弦相似度的计算

```python
def calculate_cosine_sim(x:list[float],y:list[float]):
    """
    This function use cosine sim to calculate two embeddings result sim.
    """
    dot_product = sum([x_*y_ for x_, y_ in zip(x,y)])
    x_scale = sum([each * each for each in x]) ** 0.5
    y_scale = sum([each * each for each in y]) ** 0.5
    return dot_product/(x_scale * y_scale)
```

## 一个简单的问答系统

```python
# 如下是零散知识
knowledge_1 = "Qwen/Qwen2-72B-Instruct 模型价格是 ¥4.13/1M tokens。"
knowledge_2 = "Qwen/Qwen2-57B-A14B-Instruct 模型价格是 ¥4.13/1M tokens。"
knowledge_3 = "Qwen/Qwen1.5-110B-Chat 模型价格是 ¥4.13/1M tokens。"
knowledge_4 = "Qwen/Qwen1.5-32B-Chat 模型价格是 ¥1.26/1M tokens。"
knowledge_5 = "Qwen/Qwen1.5-14B-Chat 模型价格是 ¥0.70/1M tokens。"
knowledge_6 = "deepseek-ai/DeepSeek-Coder-V2-Instruct 模型价格是 ¥1.33/1M tokens。"
knowledge_7 = "deepseek-ai/deepseek-llm-67b-chat 模型价格是 ¥1.00/1M tokens。"

knowldges = [knowledge_1,knowledge_2,knowledge_3,knowledge_4,knowledge_5,knowledge_6,knowledge_7]

# 以下是询问语句
querry_1 = "QWen2模型的价格是多少？"
querry_2 = "deepseek-coder 模型价格是多少？"

querries = [querry_1, querry_2]

knowledge_embeddings = []
qurry_embbedings = []
```

```python
def get_embedding_result(text):
    payload = {
    "model": "BAAI/bge-large-zh-v1.5",
    "input": f"{text}"
    }
    headers = {
        "accept": "application/json",
        "content-type": "application/json",
        "authorization": f"Bearer {OPENAI_KEY}"
    }

    res = requests.post(embedding_url, json=payload, headers=headers)
    return eval(res.text)['data'][0]['embedding']
```

## 生成对应的embedding结果

```python
import time
## 生成querry embedding 结果
qurry_embedding_start = time.time()

for each in querries:
    qurry_embbedings.append(get_embedding_result(each))

querry_embedding_end = time.time()
print(f"Querry Size: {len(querries)} embedding cost {querry_embedding_end - qurry_embedding_start} secondes.")
```

```python
import time
## 生成 knowledge embedding 结果
knowledge_embedding_start = time.time()
knowledge_embeddings = []

for i,each in enumerate(knowldges):
    knowledge_embeddings.append(get_embedding_result(each))
    print(f"Knowledge {i + 1} embedding successful. Sleep 10 seconds to for RPS limit.")
    time.sleep(10)

knowledge_embedding_end = time.time()
print(f"Knowledge Size: {len(knowldges)} embedding cost {knowledge_embedding_end - knowledge_embedding_start - len(knowldges) * 10} secondes.")
```

### 查询querry-knowledge中的相似度

```python
results = []
for i,querry in enumerate(qurry_embbedings):
    querry_result = []
    print(f"Querry {i} sim:", end="| ")
    for knowledge in knowledge_embeddings:
        
        sim_res = calculate_cosine_sim(querry,knowledge) 
        querry_result.append(sim_res)
        print(sim_res, end='|')
    results.append(querry_result)
    print("\n")
```

### Get top K answer as prompt

```python
top_k = 5
knowledge_from_top_k = []
for querry_sim in results:
    knowledge_from_top_k.append(sorted(range(len(querry_sim)), key=lambda i: querry_sim[i], reverse=True)[:top_k])
```

### Prompt Generate

```python
def generate_info_prompt(top_K:list[int]):
    res = ''
    '''
    Based on known indices of knowledge to generate background knowledge.
    '''
    for each in top_K:
        res = res + knowldges[each] +'\n'
    
    prompt = f'''
    已知如下信息：{res},请用简洁的语言回答下面问题：
    '''
    return prompt
```

```python
chat_postfix = r"/v1/chat/completions"
chat_url = BASE_URL + chat_postfix

payload = {
    "model": "Qwen/Qwen2-7B-Instruct",
    "messages": [
        {
            "role": "user",
            "content": f"{generate_info_prompt(knowledge_from_top_k[0])}+{querry_1}"
        }
    ]
}
headers = {
    "accept": "application/json",
    "content-type": "application/json",
    "authorization": f"Bearer {OPENAI_KEY}"
}

response = requests.post(chat_url, json=payload, headers=headers)

print(response.text)
```

```python
print(f"Question: {querries[0]}")
print("GPT with Vector Search anwser: ")
print(eval(response.text)['choices'][0]['message']['content'])
print("Correct Anwser:")
print(knowldges[:5])
```
